# Transformasi Linear

Diketahui:

$$
T(x,y)=(3x-2y,\;x+4y)
$$

Tentukan:

$$
T(2,1)
$$

## Penyelesaian

Substitusi:

$$
x=2
$$

$$
y=1
$$

$$
T(2,1)
=
(3(2)-2(1),\;2+4(1))
$$

$$
=
(6-2,\;2+4)
$$

$$
=
(4,\;6)
$$

Jadi:

$$
\boxed{T(2,1)=(4,6)}
$$

---

# Matriks Standar Transformasi Linear

Diketahui:

$$
T(x,y)=(3x-2y,\;x+4y)
$$

## Langkah 1

Basis standar:

$$
e_1=(1,0)
$$

$$
e_2=(0,1)
$$

## Langkah 2

Hitung:

$$
T(1,0)
=
(3(1)-2(0),\;1+4(0))
$$

$$
=
(3,1)
$$

Hitung:

$$
T(0,1)
=
(3(0)-2(1),\;0+4(1))
$$

$$
=
(-2,4)
$$

## Langkah 3

Susun menjadi matriks standar:

$$
A=[T(e_1)\;T(e_2)]
$$

$$
A=
\begin{bmatrix}
3 & -2\\
1 & 4
\end{bmatrix}
$$

Jadi:

$$
\boxed{
A=
\begin{bmatrix}
3 & -2\\
1 & 4
\end{bmatrix}
}
$$

$$
A=
\begin{bmatrix}

\end{bmatrix}
$$

## Langkah 1

Ambil kolom-kolom matriks:

$$

\begin{bmatrix}

\end{bmatrix}

$$

\begin{bmatrix}

\end{bmatrix}

=15

# Dekomposisi QR - Matriks 2×2

## Matriks Awal

$$
A = \begin{bmatrix} 3 & 4 \\ 0 & 5 \end{bmatrix}
$$

Kolom-kolom matriks A:

$$
a_1 = \begin{bmatrix} 3 \\ 0 \end{bmatrix}, \quad a_2 = \begin{bmatrix} 4 \\ 5 \end{bmatrix}
$$

---

## Proses Gram-Schmidt

### Langkah 1: Hitung norm dari \( a_1 \)

$$
\|a_1\| = \sqrt{3^2 + 0^2} = \sqrt{9} = 3
$$

---

### Langkah 2: Hitung \( q_1 \) (kolom pertama Q)

$$
q_1 = \frac{a_1}{\|a_1\|} = \frac{1}{3} \begin{bmatrix} 3 \\ 0 \end{bmatrix} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}
$$

---

### Langkah 3: Simpan \( R_{11} \)

$$
R_{11} = \|a_1\| = 3
$$

---

### Langkah 4: Hitung proyeksi \( a_2 \) ke \( q_1 \)

$$
a_2 \cdot q_1 = \begin{bmatrix} 4 \\ 5 \end{bmatrix} \cdot \begin{bmatrix} 1 \\ 0 \end{bmatrix} = (4 \times 1) + (5 \times 0) = 4
$$

Maka:

$$
R_{12} = 4
$$

---

### Langkah 5: Hitung \( v_2 \) (vektor tegak lurus)

$$
v_2 = a_2 - (a_2 \cdot q_1) q_1
$$

$$
v_2 = \begin{bmatrix} 4 \\ 5 \end{bmatrix} - 4 \begin{bmatrix} 1 \\ 0 \end{bmatrix}
= \begin{bmatrix} 4 \\ 5 \end{bmatrix} - \begin{bmatrix} 4 \\ 0 \end{bmatrix}
= \begin{bmatrix} 0 \\ 5 \end{bmatrix}
$$

---

### Langkah 6: Hitung norm dari \( v_2 \)

$$
\|v_2\| = \sqrt{0^2 + 5^2} = \sqrt{25} = 5
$$

Maka:

$$
R_{22} = 5
$$

---

### Langkah 7: Hitung \( q_2 \) (kolom kedua Q)

$$
q_2 = \frac{v_2}{\|v_2\|} = \frac{1}{5} \begin{bmatrix} 0 \\ 5 \end{bmatrix} = \begin{bmatrix} 0 \\ 1 \end{bmatrix}
$$

---

## Hasil Akhir

### Matriks Q

$$
Q = \begin{bmatrix} q_1 & q_2 \end{bmatrix}
= \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}
$$

### Matriks R

$$
R = \begin{bmatrix} R_{11} & R_{12} \\ 0 & R_{22} \end{bmatrix}
= \begin{bmatrix} 3 & 4 \\ 0 & 5 \end{bmatrix}
$$

---

## Verifikasi

Periksa apakah \( A = Q \times R \):

$$
Q \times R = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}
\begin{bmatrix} 3 & 4 \\ 0 & 5 \end{bmatrix}
$$

Hitung perkalian:

$$
Q \times R = \begin{bmatrix}
(1 \times 3) + (0 \times 0) & (1 \times 4) + (0 \times 5) \\
(0 \times 3) + (1 \times 0) & (0 \times 4) + (1 \times 5)
\end{bmatrix}
$$

$$
Q \times R = \begin{bmatrix} 3 & 4 \\ 0 & 5 \end{bmatrix} = A
$$

✅ **Terbukti!** \( A = QR \)
