# Transformasi Linear

Diketahui:

$$
T(x,y)=(3x-2y,\;x+4y)
$$

Tentukan:

$$
T(2,1)
$$

## Penyelesaian

Substitusi:

$$
x=2
$$

$$
y=1
$$

$$
T(2,1)
=
(3(2)-2(1),\;2+4(1))
$$

$$
=
(6-2,\;2+4)
$$

$$
=
(4,\;6)
$$

Jadi:

$$
\boxed{T(2,1)=(4,6)}
$$

---

# Matriks Standar Transformasi Linear

Diketahui:

$$
T(x,y)=(3x-2y,\;x+4y)
$$

## Langkah 1

Basis standar:

$$
e_1=(1,0)
$$

$$
e_2=(0,1)
$$

## Langkah 2

Hitung:

$$
T(1,0)
=
(3(1)-2(0),\;1+4(0))
$$

$$
=
(3,1)
$$

Hitung:

$$
T(0,1)
=
(3(0)-2(1),\;0+4(1))
$$

$$
=
(-2,4)
$$

## Langkah 3

Susun menjadi matriks standar:

$$
A=[T(e_1)\;T(e_2)]
$$

$$
A=
\begin{bmatrix}
3 & -2\\
1 & 4
\end{bmatrix}
$$

Jadi:

$$
\boxed{
A=
\begin{bmatrix}
3 & -2\\
1 & 4
\end{bmatrix}
}
$$

# Dekomposisi QR

Diketahui:

$$
A=
\begin{bmatrix}
3 & 1\\
4 & 3
\end{bmatrix}
$$

## Langkah 1

Ambil kolom-kolom matriks:

$$
a_1=
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

$$
a_2=
\begin{bmatrix}
1\\
3
\end{bmatrix}
$$

## Langkah 2

Tentukan:

$$
u_1=a_1
$$

$$
u_1=
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

## Langkah 3

Hitung norma:

$$
\|u_1\|
=
\sqrt{3^2+4^2}
$$

$$
=
\sqrt{9+16}
$$

$$
=
\sqrt{25}
$$

$$
=5
$$

## Langkah 4

Cari:

$$
e_1=
\frac{u_1}{\|u_1\|}
$$

$$
=
\frac{1}{5}
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

$$
=
\begin{bmatrix}
3/5\\
4/5
\end{bmatrix}
$$

## Langkah 5

Hitung:

$$
a_2 \cdot u_1
=
(1)(3)+(3)(4)
$$

$$
=
3+12
$$

$$
=15
$$

Hitung:

$$
u_1 \cdot u_1
=
3^2+4^2
$$

$$
=
9+16
$$

$$
=25
$$

## Langkah 6

Cari proyeksi:

$$
\operatorname{proj}_{u_1}(a_2)
=
\frac{a_2 \cdot u_1}{u_1 \cdot u_1}u_1
$$

$$
=
\frac{15}{25}
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

$$
=
\frac{3}{5}
\begin{bmatrix}
3\\
4
\end{bmatrix}
$$

$$
=
\begin{bmatrix}
9/5\\
12/5
\end{bmatrix}
$$

## Langkah 7

Cari:

$$
u_2
=
a_2-\operatorname{proj}_{u_1}(a_2)
$$

$$
=
\begin{bmatrix}
1\\
3
\end{bmatrix}
-
\begin{bmatrix}
9/5\\
12/5
\end{bmatrix}
$$

$$
=
\begin{bmatrix}
5/5-9/5\\
15/5-12/5
\end{bmatrix}
$$

$$
=
\begin{bmatrix}
-4/5\\
3/5
\end{bmatrix}
$$

## Langkah 8

Hitung norma:

$$
\|u_2\|
=
\sqrt{
\left(-\frac45\right)^2+
\left(\frac35\right)^2
}
$$

$$
=
\sqrt{
\frac{16}{25}+\frac{9}{25}
}
$$

$$
=
\sqrt{
\frac{25}{25}
}
$$

$$
=\sqrt{1}
$$

$$
=1
$$

## Langkah 9

Cari:

$$
e_2=
\frac{u_2}{\|u_2\|}
$$

$$
=
\frac{
\begin{bmatrix}
-4/5\\
3/5
\end{bmatrix}
}{1}
$$

$$
=
\begin{bmatrix}
-4/5\\
3/5
\end{bmatrix}
$$

## Langkah 10

Bentuk matriks Q:

$$
Q=
\begin{bmatrix}
3/5 & -4/5\\
4/5 & 3/5
\end{bmatrix}
$$

## Langkah 11

Bentuk matriks R:

$$
R=
\begin{bmatrix}
5 & 3\\
0 & 1
\end{bmatrix}
$$

## Hasil

$$
A=QR
$$

$$
Q=
\begin{bmatrix}
3/5 & -4/5\\
4/5 & 3/5
\end{bmatrix}
$$

$$
R=
\begin{bmatrix}
5 & 3\\
0 & 1
\end{bmatrix}
