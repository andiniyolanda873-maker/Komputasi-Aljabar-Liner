# Invers Matriks A

## 1. Diketahui

$$
A=
\begin{bmatrix}
2 & 1 & 1\\
1 & 2 & 1\\
1 & 1 & 2
\end{bmatrix}
$$

---

## 2. Rumus Invers Matriks

$$
A^{-1}=\frac{1}{\det(A)}\operatorname{adj}(A)
$$

Untuk mencari invers matriks, terlebih dahulu dicari:

1. Determinan Matriks A
2. Matriks Kofaktor
3. Adjoin Matriks
4. Invers Matriks

---

# 3. Determinan Matriks A

$$
\det(A)=
\begin{vmatrix}
2&1&1\\
1&2&1\\
1&1&2
\end{vmatrix}
$$

Ekspansi kofaktor pada baris pertama:

$$
\det(A)
=
2
\begin{vmatrix}
2&1\\
1&2
\end{vmatrix}
-
1
\begin{vmatrix}
1&1\\
1&2
\end{vmatrix}
+
1
\begin{vmatrix}
1&2\\
1&1
\end{vmatrix}
$$

### Minor Pertama

$$
\begin{vmatrix}
2&1\\
1&2
\end{vmatrix}
=
(2)(2)-(1)(1)
$$

$$
=4-1
$$

$$
=3
$$

### Minor Kedua

$$
\begin{vmatrix}
1&1\\
1&2
\end{vmatrix}
=
(1)(2)-(1)(1)
$$

$$
=2-1
$$

$$
=1
$$

### Minor Ketiga

$$
\begin{vmatrix}
1&2\\
1&1
\end{vmatrix}
=
(1)(1)-(2)(1)
$$

$$
=1-2
$$

$$
=-1
$$

Substitusi:

$$
\det(A)
=
2(3)-1(1)+1(-1)
$$

$$
=6-1-1
$$

$$
=4
$$

Jadi:

$$
\boxed{\det(A)=4}
$$

# 5. Adjoin Matriks

Adjoin adalah transpose dari matriks kofaktor.

$$
adj(A)=Cof(A)^T
$$

Karena matriks kofaktor simetris, maka:

$$
adj(A)
=
\begin{bmatrix}
3 & -1 & -1\\
-1 & 3 & -1\\
-1 & -1 & 3
\end{bmatrix}
$$

---

# 6. Invers Matriks

Gunakan rumus:

$$
A^{-1}
=
\frac{1}{\det(A)}
adj(A)
$$

Substitusi nilai determinan:

$$
A^{-1}
=
\frac{1}{4}
\begin{bmatrix}
3 & -1 & -1\\
-1 & 3 & -1\\
-1 & -1 & 3
\end{bmatrix}
$$

Kalikan setiap elemen dengan $\frac14$:

$$
A^{-1}
=
\begin{bmatrix}
\frac34 & -\frac14 & -\frac14\\
-\frac14 & \frac34 & -\frac14\\
-\frac14 & -\frac14 & \frac34
\end{bmatrix}
$$
