# ELIMINASI GAUSSIAN

## A. Sistem Persamaan Linear Tiga Variabel

### 1. Sistem Persamaan

$$
\begin{cases}
x+y+z=9\\
2x+y-z=-1\\
x-y+z=7
\end{cases}
$$

### 2. Matriks Augmented

$$
\left[
\begin{array}{ccc|c}
1 & 1 & 1 & 9\\
2 & 1 & -1 & -1\\
1 & -1 & 1 & 7
\end{array}
\right]
$$

### 3. Eliminasi Gaussian

#### Langkah 1

$$R_2 \leftarrow R_2-2R_1$$

$$R_3 \leftarrow R_3-R_1$$

Hasil:

$$
\left[
\begin{array}{ccc|c}
1 & 1 & 1 & 9\\
0 & -1 & -3 & -19\\
0 & -2 & 0 & -2
\end{array}
\right]
$$

#### Langkah 2

$$R_3 \leftarrow R_3-2R_2$$

Perhitungan:

$$(0,-2,0|-2)-2(0,-1,-3|-19)$$

$$=(0,0,6|36)$$

Hasil:

$$
\left[
\begin{array}{ccc|c}
1 & 1 & 1 & 9\\
0 & -1 & -3 & -19\\
0 & 0 & 6 & 36
\end{array}
\right]
$$

### 4. Substitusi Balik

$$6z=36$$

$$z=6$$

$$-y-3z=-19$$

$$-y-3(6)=-19$$

$$-y=-1$$

$$y=1$$

$$x+y+z=9$$

$$x+1+6=9$$

$$x=2$$

### 5. Hasil

$$x=2,\qquad y=1,\qquad z=6$$

### 6. Kesimpulan

Berdasarkan metode Eliminasi Gaussian, diperoleh solusi sistem persamaan linear tiga variabel:

$$\boxed{x=2,\quad y=1,\quad z=6}$$

Nilai tersebut memenuhi seluruh persamaan yang diberikan.
