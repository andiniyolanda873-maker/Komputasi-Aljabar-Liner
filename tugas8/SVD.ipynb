# Singular Value Decomposition (SVD) 3×3

---

## Matriks A

$$
A =
\begin{bmatrix}
1 & 1 & 0 \\
1 & 0 & 1 \\
0 & 1 & 1
\end{bmatrix}
$$

Keterangan: Matriks A adalah matriks simetris yang akan didekomposisi menjadi \(A = U \Sigma V^T\).

---

# 1. Transpose Matriks A

$$
A^T =
\begin{bmatrix}
1 & 1 & 0 \\
1 & 0 & 1 \\
0 & 1 & 1
\end{bmatrix}
$$

Keterangan: Karena A simetris, maka \(A^T = A\).

---

# 2. Menghitung \(A^T A\)

$$
A^T A =
\begin{bmatrix}
2 & 1 & 1 \\
1 & 2 & 1 \\
1 & 1 & 2
\end{bmatrix}
$$

Keterangan: Matriks ini selalu simetris dan digunakan untuk mencari eigenvalue pada SVD.

---

# 3. Eigenvalue dari \(A^T A\)

Gunakan:

$$
\det(A^T A - \lambda I)=0
$$

$$
\begin{vmatrix}
2-\lambda & 1 & 1 \\
1 & 2-\lambda & 1 \\
1 & 1 & 2-\lambda
\end{vmatrix}=0
$$

Hasil eigenvalue:

$$
\lambda_1 = 4,\quad \lambda_2 = 1,\quad \lambda_3 = 1
$$

Keterangan: Eigenvalue menunjukkan besarnya energi atau kontribusi setiap arah transformasi.

---

# 4. Singular Value

$$
\sigma_i = \sqrt{\lambda_i}
$$

Hasil:

$$
\sigma_1 = 2,\quad \sigma_2 = 1,\quad \sigma_3 = 1
$$

Keterangan: Singular value menunjukkan seberapa kuat suatu arah ditransformasikan.

---

# 5. Matriks Σ (Sigma)

$$
\Sigma =
\begin{bmatrix}
2 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

Keterangan: Matriks diagonal berisi singular value dari besar ke kecil.

---

# 6. Matriks V (Right Singular Vector)

## Untuk λ = 4

$$
v_1 =
\frac{1}{\sqrt{3}}
\begin{bmatrix}
1 \\
1 \\
1
\end{bmatrix}
$$

## Untuk λ = 1

$$
v_2 =
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 \\
-1 \\
0
\end{bmatrix}
$$

$$
v_3 =
\frac{1}{\sqrt{6}}
\begin{bmatrix}
1 \\
1 \\
-2
\end{bmatrix}
$$

Keterangan: V menunjukkan arah utama sebelum transformasi.

---

## Matriks V

$$
V =
\begin{bmatrix}
\frac{1}{\sqrt{3}} & \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{6}} \\
\frac{1}{\sqrt{3}} & -\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{6}} \\
\frac{1}{\sqrt{3}} & 0 & -\frac{2}{\sqrt{6}}
\end{bmatrix}
$$

---

## Transpose V

$$
V^T =
\begin{bmatrix}
\frac{1}{\sqrt{3}} & \frac{1}{\sqrt{3}} & \frac{1}{\sqrt{3}} \\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} & 0 \\
\frac{1}{\sqrt{6}} & \frac{1}{\sqrt{6}} & -\frac{2}{\sqrt{6}}
\end{bmatrix}
$$

Keterangan: \(V^T\) digunakan dalam rekonstruksi matriks A.

---

# 7. Matriks U (Left Singular Vector)

Gunakan rumus:

$$
u_i = \frac{A v_i}{\sigma_i}
$$

---

## u₁

$$
A v_1 =
\begin{bmatrix}
2 \\
2 \\
2
\end{bmatrix}
$$

$$
u_1 =
\frac{1}{\sqrt{3}}
\begin{bmatrix}
1 \\
1 \\
1
\end{bmatrix}
$$

---

## u₂

$$
A v_2 =
\begin{bmatrix}
1 \\
-1 \\
0
\end{bmatrix}
$$

$$
u_2 =
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 \\
-1 \\
0
\end{bmatrix}
$$

---

## u₃

$$
A v_3 =
\begin{bmatrix}
1 \\
1 \\
-2
\end{bmatrix}
$$

$$
u_3 =
\frac{1}{\sqrt{6}}
\begin{bmatrix}
1 \\
1 \\
-2
\end{bmatrix}
$$

---

## Matriks U

$$
U =
\begin{bmatrix}
\frac{1}{\sqrt{3}} & \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{6}} \\
\frac{1}{\sqrt{3}} & -\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{6}} \\
\frac{1}{\sqrt{3}} & 0 & -\frac{2}{\sqrt{6}}
\end{bmatrix}
$$

Keterangan: U menunjukkan arah hasil setelah transformasi.

---

# 8. Hasil Akhir SVD

$$
A = U \Sigma V^T
$$

Keterangan: Matriks A dapat dipisahkan menjadi rotasi → skala → rotasi.

---

# 9. Verifikasi Python (opsional)

```python
import numpy as np

A = np.array([[1,1,0],
              [1,0,1],
              [0,1,1]], dtype=float)

U, s, Vt = np.linalg.svd(A)

Sigma = np.zeros((3,3))
np.fill_diagonal(Sigma, s)

A_rekonstruksi = U @ Sigma @ Vt

print(A_rekonstruksi)
