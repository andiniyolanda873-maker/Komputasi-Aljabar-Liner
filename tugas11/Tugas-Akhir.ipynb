# A. SOAL DETERMINAN MATRIKS 3×3

---

## Soal 1

Tentukan determinan dari matriks berikut:

$$
A = \begin{bmatrix}
2 & 1 & 3 \\
0 & 4 & 5 \\
1 & 2 & 1
\end{bmatrix}
$$

### Penyelesaian

**Rumus determinan matriks 3×3:**

Untuk matriks:

$$
M = \begin{bmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{bmatrix}
$$

maka:

$$
\det(M) = a(ei - fh) - b(di - fg) + c(dh - eg)
$$

**Identifikasi elemen:**

$$
a = 2,\ b = 1,\ c = 3
$$

$$
d = 0,\ e = 4,\ f = 5
$$

$$
g = 1,\ h = 2,\ i = 1
$$

**Substitusi ke rumus:**

$$
\det(A) = 2(4 \times 1 - 5 \times 2) - 1(0 \times 1 - 5 \times 1) + 3(0 \times 2 - 4 \times 1)
$$

$$
= 2(4 - 10) - 1(0 - 5) + 3(0 - 4)
$$

$$
= 2(-6) - 1(-5) + 3(-4)
$$

$$
= -12 + 5 - 12
$$

$$
= -19
$$

**Jawaban:**

$$
\boxed{\det(A) = -19}
$$

## Soal 2

Tentukan determinan dari matriks berikut:

$$
B = \begin{bmatrix}
3 & 2 & 1 \\
1 & 0 & 4 \\
2 & 5 & 1
\end{bmatrix}
$$

### Penyelesaian

**Identifikasi elemen:**

$$
a = 3,\ b = 2,\ c = 1
$$

$$
d = 1,\ e = 0,\ f = 4
$$

$$
g = 2,\ h = 5,\ i = 1
$$

**Substitusi ke rumus:**

$$
\det(B) = 3(0 \times 1 - 4 \times 5) - 2(1 \times 1 - 4 \times 2) + 1(1 \times 5 - 0 \times 2)
$$

$$
= 3(0 - 20) - 2(1 - 8) + 1(5 - 0)
$$

$$
= 3(-20) - 2(-7) + 5
$$

$$
= -60 + 14 + 5
$$

$$
= -41
$$

**Jawaban:**

$$
\boxed{\det(B) = -41}
$$

## Soal 3

Tentukan determinan dari matriks berikut:

$$
C = \begin{bmatrix}
1 & 2 & 3 \\
2 & 4 & 6 \\
1 & 1 & 1
\end{bmatrix}
$$

### Penyelesaian

**Identifikasi elemen:**

$$
a = 1,\ b = 2,\ c = 3
$$

$$
d = 2,\ e = 4,\ f = 6
$$

$$
g = 1,\ h = 1,\ i = 1
$$

**Substitusi ke rumus:**

$$
\det(C) = 1(4 \times 1 - 6 \times 1) - 2(2 \times 1 - 6 \times 1) + 3(2 \times 1 - 4 \times 1)
$$

$$
= 1(4 - 6) - 2(2 - 6) + 3(2 - 4)
$$

$$
= 1(-2) - 2(-4) + 3(-2)
$$

$$
= -2 + 8 - 6
$$

$$
= 0
$$

**Jawaban:**

$$
\boxed{\det(C) = 0}
$$

**Kesimpulan:** Matriks C **singular** (tidak memiliki invers) karena determinannya = 0.

# B. SOAL DEKOMPOSISI LU

---

## Soal 4

Lakukan dekomposisi LU terhadap matriks:

$$
A = \begin{bmatrix}
2 & 4 & 2 \\
1 & 5 & 2 \\
1 & 2 & 4
\end{bmatrix}
$$

dengan bentuk \( A = LU \), di mana \( L \) matriks segitiga bawah dan \( U \) matriks segitiga atas.

### Penyelesaian

**Metode:** Eliminasi Gauss, catat pengali \( m_{ij} \)

**Langkah 1:** Pivot baris 1 kolom 1 = 2

$$
m_{21} = \frac{a_{21}}{a_{11}} = \frac{1}{2}
$$

Baris 2 baru:

$$
[1,5,2] - \frac{1}{2}[2,4,2]
$$

$$
= [1-1,\ 5-2,\ 2-1]
$$

$$
= [0,3,1]
$$

**Langkah 2:** 

$$
m_{31} = \frac{a_{31}}{a_{11}} = \frac{1}{2}
$$

Baris 3 baru:

$$
[1,2,4] - \frac{1}{2}[2,4,2]
$$

$$
= [1-1,\ 2-2,\ 4-1]
$$

$$
= [0,0,3]
$$

**Langkah 3:** Pivot baris 2 kolom 2 = 3

$$
m_{32} = \frac{a_{32}}{a_{22}} = \frac{0}{3} = 0
$$

Baris 3 baru (tetap):

$$
[0,0,3]
$$

**Hasil:**

$$
U = \begin{bmatrix}
2 & 4 & 2 \\
0 & 3 & 1 \\
0 & 0 & 3
\end{bmatrix}
$$

$$
L = \begin{bmatrix}
1 & 0 & 0 \\
\frac{1}{2} & 1 & 0 \\
\frac{1}{2} & 0 & 1
\end{bmatrix}
$$

**Verifikasi \( A = LU \):**

$$
L \times U = 
\begin{bmatrix}
1 & 0 & 0 \\
\frac{1}{2} & 1 & 0 \\
\frac{1}{2} & 0 & 1
\end{bmatrix}
\begin{bmatrix}
2 & 4 & 2 \\
0 & 3 & 1 \\
0 & 0 & 3
\end{bmatrix}
$$

$$
= \begin{bmatrix}
1(2)+0+0 & 1(4)+0+0 & 1(2)+0+0 \\
\frac{1}{2}(2)+1(0)+0 & \frac{1}{2}(4)+1(3)+0 & \frac{1}{2}(2)+1(1)+0 \\
\frac{1}{2}(2)+0(0)+0 & \frac{1}{2}(4)+0(3)+0(0) & \frac{1}{2}(2)+0(1)+1(3)
\end{bmatrix}
$$

$$
= \begin{bmatrix}
2 & 4 & 2 \\
1 & 5 & 2 \\
1 & 2 & 4
\end{bmatrix}
= A
$$

**Jawaban:**

$$
\boxed{
L = \begin{bmatrix}
1 & 0 & 0 \\
\frac{1}{2} & 1 & 0 \\
\frac{1}{2} & 0 & 1
\end{bmatrix},\quad
U = \begin{bmatrix}
2 & 4 & 2 \\
0 & 3 & 1 \\
0 & 0 & 3
\end{bmatrix}
}
$$

## Soal 5

Tentukan matriks \( L \) dan \( U \) dari:

$$
B = \begin{bmatrix}
1 & 2 & 1 \\
2 & 5 & 3 \\
4 & 10 & 8
\end{bmatrix}
$$

### Penyelesaian

**Langkah 1:** Pivot baris 1 kolom 1 = 1

$$
m_{21} = \frac{2}{1} = 2
$$

Baris 2 baru:

$$
[2,5,3] - 2[1,2,1]
$$

$$
= [2-2,\ 5-4,\ 3-2]
$$

$$
= [0,1,1]
$$

**Langkah 2:**

$$
m_{31} = \frac{4}{1} = 4
$$

Baris 3 baru:

$$
[4,10,8] - 4[1,2,1]
$$

$$
= [4-4,\ 10-8,\ 8-4]
$$

$$
= [0,2,4]
$$

**Langkah 3:** Pivot baris 2 kolom 2 = 1

$$
m_{32} = \frac{2}{1} = 2
$$

Baris 3 baru:

$$
[0,2,4] - 2[0,1,1]
$$

$$
= [0,\ 2-2,\ 4-2]
$$

$$
= [0,0,2]
$$

**Hasil:**

$$
U = \begin{bmatrix}
1 & 2 & 1 \\
0 & 1 & 1 \\
0 & 0 & 2
\end{bmatrix}
$$

$$
L = \begin{bmatrix}
1 & 0 & 0 \\
2 & 1 & 0 \\
4 & 2 & 1
\end{bmatrix}
$$

**Verifikasi \( B = LU \):**

$$
L \times U = 
\begin{bmatrix}
1 & 0 & 0 \\
2 & 1 & 0 \\
4 & 2 & 1
\end{bmatrix}
\begin{bmatrix}
1 & 2 & 1 \\
0 & 1 & 1 \\
0 & 0 & 2
\end{bmatrix}
$$

$$
= \begin{bmatrix}
1 & 2 & 1 \\
2 & 5 & 3 \\
4 & 10 & 8
\end{bmatrix}
= B
$$

**Jawaban:**

$$
\boxed{
L = \begin{bmatrix}
1 & 0 & 0 \\
2 & 1 & 0 \\
4 & 2 & 1
\end{bmatrix},\quad
U = \begin{bmatrix}
1 & 2 & 1 \\
0 & 1 & 1 \\
0 & 0 & 2
\end{bmatrix}
}
$$

## Soal 6

Dekomposisikan matriks berikut menjadi \( LU \):

$$
C = \begin{bmatrix}
4 & 2 & 0 \\
2 & 5 & 1 \\
0 & 1 & 3
\end{bmatrix}
$$

### Penyelesaian

**Langkah 1:** Pivot baris 1 kolom 1 = 4

$$
m_{21} = \frac{2}{4} = \frac{1}{2}
$$

Baris 2 baru:

$$
[2,5,1] - \frac{1}{2}[4,2,0]
$$

$$
= [2-2,\ 5-1,\ 1-0]
$$

$$
= [0,4,1]
$$

**Langkah 2:**

$$
m_{31} = \frac{0}{4} = 0
$$

Baris 3 tetap:

$$
[0,1,3]
$$

**Langkah 3:** Pivot baris 2 kolom 2 = 4

$$
m_{32} = \frac{1}{4}
$$

Baris 3 baru:

$$
[0,1,3] - \frac{1}{4}[0,4,1]
$$

$$
= [0,\ 1-1,\ 3-\frac{1}{4}]
$$

$$
= [0,\ 0,\ \frac{12}{4}-\frac{1}{4}]
$$

$$
= [0,\ 0,\ \frac{11}{4}]
$$

**Hasil:**

$$
U = \begin{bmatrix}
4 & 2 & 0 \\
0 & 4 & 1 \\
0 & 0 & \frac{11}{4}
\end{bmatrix}
$$

$$
L = \begin{bmatrix}
1 & 0 & 0 \\
\frac{1}{2} & 1 & 0 \\
0 & \frac{1}{4} & 1
\end{bmatrix}
$$

**Verifikasi \( C = LU \):**

$$
L \times U = 
\begin{bmatrix}

0 & \frac{1}{4} & 1
\end{bmatrix}
\begin{bmatrix}
4 & 2 & 0 \\
0 & 4 & 1 \\
0 & 0 & \frac{11}{4}
\end{bmatrix}

$$
= \begin{bmatrix}
4 & 2 & 0 \\
2 & 5 & 1 \\

**Jawaban:**

$$
\boxed{
L = \begin{bmatrix}
1 & 0 & 0 \\
\frac{1}{2} & 1 & 0 \\
0 & \frac{1}{4} & 1
\end{bmatrix},\quad

4 & 2 & 0 \\
0 & 4 & 1 \\
0 & 0 & \frac{11}{4}
\end{bmatrix}
}

# C. SOAL INVERS MATRIKS 3×3

---

## Soal 7

Tentukan invers dari matriks:

$$
A = \begin{bmatrix}
1 & 2 & 1 \\
0 & 1 & 1 \\
2 & 3 & 4
\end{bmatrix}
$$

### Penyelesaian

**Rumus Invers:**

$$
A^{-1} = \frac{1}{\det(A)} \times \text{adj}(A)
$$

di mana:

$$
\text{adj}(A) = \text{transpose dari matriks kofaktor}
$$

**Langkah 1: Hitung determinan**

$$
\det(A) = 1(1 \times 4 - 1 \times 3) - 2(0 \times 4 - 1 \times 2) + 1(0 \times 3 - 1 \times 2)
$$

$$
= 1(4 - 3) - 2(0 - 2) + 1(0 - 2)
$$

$$
= 1(1) - 2(-2) + 1(-2)
$$

$$
= 1 + 4 - 2
$$

$$
= 3
$$

**Langkah 2: Hitung kofaktor**

$$
C_{11} = +\begin{vmatrix}1 & 1 \\ 3 & 4\end{vmatrix} = 1 \times 4 - 1 \times 3 = 4 - 3 = 1
$$

$$
C_{12} = -\begin{vmatrix}0 & 1 \\ 2 & 4\end{vmatrix} = -(0 \times 4 - 1 \times 2) = -(0 - 2) = 2
$$

$$
C_{13} = +\begin{vmatrix}0 & 1 \\ 2 & 3\end{vmatrix} = 0 \times 3 - 1 \times 2 = 0 - 2 = -2
$$

$$
C_{21} = -\begin{vmatrix}2 & 1 \\ 3 & 4\end{vmatrix} = -(2 \times 4 - 1 \times 3) = -(8 - 3) = -5
$$

$$
C_{22} = +\begin{vmatrix}1 & 1 \\ 2 & 4\end{vmatrix} = 1 \times 4 - 1 \times 2 = 4 - 2 = 2
$$

$$
C_{23} = -\begin{vmatrix}1 & 2 \\ 2 & 3\end{vmatrix} = -(1 \times 3 - 2 \times 2) = -(3 - 4) = 1
$$

$$
C_{31} = +\begin{vmatrix}2 & 1 \\ 1 & 1\end{vmatrix} = 2 \times 1 - 1 \times 1 = 2 - 1 = 1
$$

$$
C_{32} = -\begin{vmatrix}1 & 1 \\ 0 & 1\end{vmatrix} = -(1 \times 1 - 1 \times 0) = -(1 - 0) = -1
$$

$$
C_{33} = +\begin{vmatrix}1 & 2 \\ 0 & 1\end{vmatrix} = 1 \times 1 - 2 \times 0 = 1 - 0 = 1
$$

**Matriks kofaktor:**

$$
\text{Cof}(A) = \begin{bmatrix}
1 & 2 & -2 \\
-5 & 2 & 1 \\
1 & -1 & 1
\end{bmatrix}
$$

**Langkah 3: Adjoin (transpose dari kofaktor)**

$$
\text{adj}(A) = \begin{bmatrix}
1 & -5 & 1 \\
2 & 2 & -1 \\
-2 & 1 & 1
\end{bmatrix}
$$

**Langkah 4: Invers**

$$
A^{-1} = \frac{1}{3}
\begin{bmatrix}
1 & -5 & 1 \\
2 & 2 & -1 \\
-2 & 1 & 1
\end{bmatrix}
$$

$$
A^{-1} = \begin{bmatrix}
\frac{1}{3} & -\frac{5}{3} & \frac{1}{3} \\
\frac{2}{3} & \frac{2}{3} & -\frac{1}{3} \\
-\frac{2}{3} & \frac{1}{3} & \frac{1}{3}
\end{bmatrix}
$$

**Jawaban:**

$$
\boxed{
A^{-1} = \frac{1}{3}
\begin{bmatrix}
1 & -5 & 1 \\
2 & 2 & -1 \\
-2 & 1 & 1
\end{bmatrix}
}
$$

## Soal 8

Carilah invers matriks:

$$
B = \begin{bmatrix}
2 & 1 & 0 \\
1 & 2 & 1 \\
0 & 1 & 2
\end{bmatrix}
$$

### Penyelesaian

**Langkah 1: Hitung determinan**

$$
\det(B) = 2(2 \times 2 - 1 \times 1) - 1(1 \times 2 - 1 \times 0) + 0
$$

$$
= 2(4 - 1) - 1(2 - 0)
$$

$$
= 2(3) - 2
$$

$$
= 6 - 2
$$

$$
= 4
$$

**Langkah 2: Hitung kofaktor**

$$
C_{11} = +\begin{vmatrix}2 & 1 \\ 1 & 2\end{vmatrix} = 2 \times 2 - 1 \times 1 = 4 - 1 = 3
$$

$$
C_{12} = -\begin{vmatrix}1 & 1 \\ 0 & 2\end{vmatrix} = -(1 \times 2 - 1 \times 0) = -(2 - 0) = -2
$$

$$
C_{13} = +\begin{vmatrix}1 & 2 \\ 0 & 1\end{vmatrix} = 1 \times 1 - 2 \times 0 = 1 - 0 = 1
$$

$$
C_{21} = -\begin{vmatrix}1 & 0 \\ 1 & 2\end{vmatrix} = -(1 \times 2 - 0 \times 1) = -(2 - 0) = -2
$$

$$
C_{22} = +\begin{vmatrix}2 & 0 \\ 0 & 2\end{vmatrix} = 2 \times 2 - 0 \times 0 = 4 - 0 = 4
$$

$$
C_{23} = -\begin{vmatrix}2 & 1 \\ 0 & 1\end{vmatrix} = -(2 \times 1 - 1 \times 0) = -(2 - 0) = -2
$$

$$
C_{31} = +\begin{vmatrix}1 & 0 \\ 2 & 1\end{vmatrix} = 1 \times 1 - 0 \times 2 = 1 - 0 = 1
$$

$$
C_{32} = -\begin{vmatrix}2 & 0 \\ 1 & 1\end{vmatrix} = -(2 \times 1 - 0 \times 1) = -(2 - 0) = -2
$$

$$
C_{33} = +\begin{vmatrix}2 & 1 \\ 1 & 2\end{vmatrix} = 2 \times 2 - 1 \times 1 = 4 - 1 = 3
$$

**Matriks kofaktor:**

$$
\text{Cof}(B) = \begin{bmatrix}
3 & -2 & 1 \\
-2 & 4 & -2 \\
1 & -2 & 3
\end{bmatrix}
$$

**Langkah 3: Adjoin (transpose)**

Karena matriks simetris, maka:

$$
\text{adj}(B) = \begin{bmatrix}
3 & -2 & 1 \\
-2 & 4 & -2 \\
1 & -2 & 3
\end{bmatrix}
$$

**Langkah 4: Invers**

$$
B^{-1} = \frac{1}{4}
\begin{bmatrix}
3 & -2 & 1 \\
-2 & 4 & -2 \\
1 & -2 & 3
\end{bmatrix}
$$

$$
B^{-1} = \begin{bmatrix}
\frac{3}{4} & -\frac{1}{2} & \frac{1}{4} \\
-\frac{1}{2} & 1 & -\frac{1}{2} \\
\frac{1}{4} & -\frac{1}{2} & \frac{3}{4}
\end{bmatrix}
$$

**Jawaban:**

$$
\boxed{
B^{-1} = \frac{1}{4}
\begin{bmatrix}
3 & -2 & 1 \\
-2 & 4 & -2 \\
1 & -2 & 3
\end{bmatrix}
}
$$

## Soal 9

Diketahui:

$$
C = \begin{bmatrix}
3 & 0 & 2 \\
2 & 0 & -2 \\
0 & 1 & 1
\end{bmatrix}
$$

Tentukan \( C^{-1} \).

### Penyelesaian

**Langkah 1: Hitung determinan**

$$
\det(C) = 3(0 \times 1 - (-2) \times 1) - 0(2 \times 1 - (-2) \times 0) + 2(2 \times 1 - 0 \times 0)
$$

$$
= 3(0 - (-2)) - 0 + 2(2 - 0)
$$

$$
= 3(0 + 2) + 2(2)
$$

$$
= 3(2) + 4
$$

$$
= 6 + 4
$$

$$
= 10
$$

**Langkah 2: Hitung kofaktor**

$$
C_{11} = +\begin{vmatrix}0 & -2 \\ 1 & 1\end{vmatrix} = 0 \times 1 - (-2) \times 1 = 0 + 2 = 2
$$

$$
C_{12} = -\begin{vmatrix}2 & -2 \\ 0 & 1\end{vmatrix} = -(2 \times 1 - (-2) \times 0) = -(2 - 0) = -2
$$

$$
C_{13} = +\begin{vmatrix}2 & 0 \\ 0 & 1\end{vmatrix} = 2 \times 1 - 0 \times 0 = 2 - 0 = 2
$$

$$
C_{21} = -\begin{vmatrix}0 & 2 \\ 1 & 1\end{vmatrix} = -(0 \times 1 - 2 \times 1) = -(0 - 2) = 2
$$

$$
C_{22} = +\begin{vmatrix}3 & 2 \\ 0 & 1\end{vmatrix} = 3 \times 1 - 2 \times 0 = 3 - 0 = 3
$$

$$
C_{23} = -\begin{vmatrix}3 & 0 \\ 0 & 1\end{vmatrix} = -(3 \times 1 - 0 \times 0) = -(3 - 0) = -3
$$

$$
C_{31} = +\begin{vmatrix}0 & 2 \\ 0 & -2\end{vmatrix} = 0 \times (-2) - 2 \times 0 = 0 - 0 = 0
$$

$$
C_{32} = -\begin{vmatrix}3 & 2 \\ 2 & -2\end{vmatrix} = -(3 \times (-2) - 2 \times 2) = -(-6 - 4) = -(-10) = 10
$$

$$
C_{33} = +\begin{vmatrix}3 & 0 \\ 2 & 0\end{vmatrix} = 3 \times 0 - 0 \times 2 = 0 - 0 = 0
$$

**Matriks kofaktor:**

$$
\text{Cof}(C) = \begin{bmatrix}
2 & -2 & 2 \\
2 & 3 & -3 \\
0 & 10 & 0
\end{bmatrix}
$$

**Langkah 3: Adjoin (transpose dari kofaktor)**

$$
\text{adj}(C) = \begin{bmatrix}
2 & 2 & 0 \\
-2 & 3 & 10 \\
2 & -3 & 0
\end{bmatrix}
$$

**Langkah 4: Invers**

$$
C^{-1} = \frac{1}{10}
\begin{bmatrix}
2 & 2 & 0 \\
-2 & 3 & 10 \\
2 & -3 & 0
\end{bmatrix}
$$

$$
C^{-1} = \begin{bmatrix}
\frac{1}{5} & \frac{1}{5} & 0 \\
-\frac{1}{5} & \frac{3}{10} & 1 \\
\frac{1}{5} & -\frac{3}{10} & 0
\end{bmatrix}
$$

**Jawaban:**

$$
\boxed{
C^{-1} = \frac{1}{10}
\begin{bmatrix}
2 & 2 & 0 \\
-2 & 3 & 10 \\
2 & -3 & 0
\end{bmatrix}
}
$$
